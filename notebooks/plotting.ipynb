```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Frequency')
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplotof Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot( [tips[tips['sex'] == 'Female']['tip'], tips[tips['sex'] == 'Male']['tip']],
             labels=['Female', 'Male'])
axes1.set_title('Boxplot of tips by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
plt.show()
```

```python
def recode_sex(sex):
    return 0 if sex == 'Female' else 1
        
tips['sex_color']=tips['sex'].apply(recode_sex)
tips.head()
```

```python

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(x=tips['total_bill'], y=tips['tip'],s=tips['size']*10, c=tips['sex_color'], alpha=0.5)
axes1.set_title('Scatterplotof Total Bill vs Tip by Sex and sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(x=tips[tips.sex == 'Female']['total_bill'], y=tips[tips.sex == 'Female']['tip'],s=tips['size']*10, c='r', alpha=0.5, label='Female')
axes1.scatter(x=tips[tips.sex == 'Male']['total_bill'], y=tips[tips.sex == 'Male']['tip'],s=tips['size']*10, c='b', alpha=0.5, label='Male')
axes1.set_title('Scatterplotof Total Bill vs Tip by Sex and sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
axes1.legend()
```
